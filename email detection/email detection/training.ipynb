```python
import pandas as pd
import joblib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# 1. Load historical and API data
df_history = pd.read_csv('spam.csv', encoding='latin-1')[['Message']]
df_live = pd.read_csv('real_time_api_data.csv')

# 2. Combine and Clean
df = pd.concat([df_history, df_live], ignore_index=True)
df['Message'] = df['Message'].fillna('').astype(str)
df = df[df['Message'].str.strip() != '']

# 3. Vectorizer (Imported correctly from feature_extraction)
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english', max_features=2500)
X = tfidf.fit_transform(df['Message'])

# 4. Unsupervised Clustering
model = KMeans(n_clusters=2, init='k-means++', n_init=10, random_state=42)
model.fit(X)

# 5. Save the Updated Brain
joblib.dump(model, 'unsupervised_model.pkl')
joblib.dump(tfidf, 'vectorizer.pkl')
print("Model re-trained with live API data!")
```
